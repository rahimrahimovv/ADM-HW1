## Birthday Cake Candles

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Define a function to count the number of tallest candles.
def birthdayCakeCandles(candles):
    # The function takes a list of candles as input.

    # The max() function finds the tallest candle by getting the maximum value in the list.
    # Then, the count() function counts how many times the tallest candle value appears in the list.
    # This gives us the number of tallest candles.

    # Return the count of tallest candles.
    return candles.count(max(candles))

if __name__ == '__main__':
    # Open a file for writing the output.
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # Read the number of candles from the input.
    candles_count = int(input().strip())

    # Read the heights of candles as a list of integers from the input.
    candles = list(map(int, input().rstrip().split()))

    # Call the 'birthdayCakeCandles' function to calculate the result.
    result = birthdayCakeCandles(candles)

    # Write the result to the output file and add a newline character.
    fptr.write(str(result) + '\n')

    # Close the output file.
    fptr.close()
```

## Number Line Jumps

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Define a function to determine if two kangaroos meet or not.
def kangaroo(x1, v1, x2, v2):
    # Check the initial conditions for the kangaroos' positions and velocities.
    
    if (v2 > v1) and (x2 > x1):
        # If the second kangaroo is faster and ahead, they will never meet.
        return "NO"
    elif v2 - v1 == 0:
        # If their velocities are the same, they will never meet.
        return "NO"
    else:
        # Calculate the time it takes for the kangaroos to meet.
        reso = (x1 - x2) % (v2 - v1)
        if reso == 0:
            # If the remainder is 0, they meet at the same position at the same time.
            return 'YES'
        else:
            # Otherwise, they don't meet at the same time and position.
            return 'NO'

if __name__ == '__main__':
    # Open a file for writing the output.
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # Read input values and parse them.
    first_multiple_input = input().rstrip().split()

    x1 = int(first_multiple_input[0])
    v1 = int(first_multiple_input[1])
    x2 = int(first_multiple_input[2])
    v2 = int(first_multiple_input[3])

    # Call the 'kangaroo' function to calculate the result.
    result = kangaroo(x1, v1, x2, v2)

    # Write the result to the output file and add a newline character.
    fptr.write(result + '\n')

    # Close the output file.
    fptr.close()
```

## Recursive Digit Sum

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Define a recursive function to calculate the super digit.
def superDigit(n, k):
    # Base case: If the number 'n' has only one digit, return that digit as an integer.
    if len(n) == 1:
        return int(n)
    else:
        # Calculate the initial digit sum and multiply it by 'k'.
        dig_init = sum(list(map(int, str(n)))
        digrep_ktimes = dig_init * k

        # Recursively call the 'superDigit' function with the new number.
        return superDigit(str(digrep_ktimes), 1)

if __name__ == '__main__':
    # Open a file for writing the output.
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # Read input values and parse them.
    first_multiple_input = input().rstrip().split()

    n = first_multiple_input[0]
    k = int(first_multiple_input[1])

    # Call the 'superDigit' function to calculate the result.
    result = superDigit(n, k)

    # Write the result to the output file and add a newline character.
    fptr.write(str(result) + '\n')

    # Close the output file.
    fptr.close()
```

## Insertion Sort - Part 1

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Define a function to perform one step of insertion sort.
def insertionSort1(n, arr):
    # Get the last element from the input array.
    last = arr[-1]
    
    # Start with the second-to-last element in the array.
    m = n - 2

    while (last < arr[m]) and (m >= 0):
        # Shift elements to the right to make space for the 'last' element.
        arr[m+1] = arr[m]
        
        # Print the current state of the array after shifting.
        print(*arr)
        
        m -= 1

    # Place the 'last' element in the correct position.
    arr[m+1] = last
    
    # Print the final state of the array after the insertion.
    print(*arr)

if __name__ == '__main__':
    # Read the number of elements 'n' from the input.
    n = int(input().strip)

    # Read the array of integers from the input.
    arr = list(map(int, input().rstrip().split()))

    # Call the 'insertionSort1' function to perform one step of insertion sort.
    insertionSort1(n, arr)
```

## Insertion Sort - Part 2

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Define a function to perform insertion sort on an array and print the intermediate steps.
def insertionSort2(n, arr):
    # Iterate over the array starting from the second element.
    for i in range(1, len(arr)):
        contest = arr[i]  # Store the current element being considered.
        v = i - 1  # Initialize a variable to track the previous element.

        # Move elements greater than the 'contest' element to the right.
        while v >= 0 and arr[v] > contest:
            arr[v+1] = arr[v]
            v = v - 1

        # Place the 'contest' element in its correct position.
        arr[v+1] = contest
        
        # Print the current state of the array after this step of insertion sort.
        print(*arr)

if __name__ == '__main__':
    # Read the number of elements 'n' from the input.
    n = int(input().strip)

    # Read the array of integers from the input.
    arr = list(map(int, input().rstrip().split()))

    # Call the 'insertionSort2' function to perform insertion sort and print intermediate steps.
    insertionSort2(n, arr)
```
